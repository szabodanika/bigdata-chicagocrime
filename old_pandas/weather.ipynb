```python
import pandas as pd
import matplotlib.pyplot as plt

weather_data = pd.read_csv('chicago_weather_2020-01-01_to_2024-04-30.csv')
weather_data['datetime'] = pd.to_datetime(weather_data['datetime'])
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(8, 4))

axes[0].hist(weather_data['temp'], bins=30, color='skyblue', edgecolor='black')
axes[0].set_title('Temperature')
axes[0].set_xlabel('Temperature (°C)')
axes[0].set_ylabel('Frequency')
axes[0].grid(True)

axes[1].hist(weather_data[weather_data['precip'] < 100]['precip'], bins=30, color='skyblue', edgecolor='black')
axes[1].set_title('Precipitation')
axes[1].set_xlabel('Precipitation (Millimeters)')
axes[1].set_ylabel('Frequency')
axes[1].set_yscale('log')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

temperature_counts = weather_data['temp'].value_counts().reset_index()
temperature_counts.columns = ['Temperature', 'Day Count']

plt.figure(figsize=(12, 6))
plt.scatter(temperature_counts['Temperature'], temperature_counts['Day Count'], alpha=0.5)
plt.title('Number of Days vs Temperature')
plt.xlabel('Temperature (°C)')
plt.ylabel('Number of Days')
plt.grid(True)
plt.tight_layout()
plt.show()
```
